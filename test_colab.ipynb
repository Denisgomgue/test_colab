```python
import numpy as np
import matplotlib.pyplot as plt
```

¿Más horas de videojuegos afectan el rendimiento académico?

```python
horas_videojuego = np.array([0, 1, 2, 3, 4, 5, 6])
notas = np.array([18, 17, 16, 14, 13, 11, 10])
```

Regresión lineal básica

```python
coef = np.polyfit(horas_videojuego, notas, 1)
prediccion = np.poly1d(coef)
```

Grafico

```python
plt.scatter(horas_videojuego, notas, color='purple')
plt.plot(horas_videojuego, prediccion(horas_videojuego), color='orange')
plt.xlabel('Horas de videojuegos al día')
plt.ylabel('Nota promedio')
plt.title('Regresión lineal: Videojuegos vs Rendimiento Académico')
plt.grid(True)
plt.savefig("regresion_numpy.png")
```
